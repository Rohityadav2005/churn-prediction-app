```python
import pickle
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
model = load_model('model.h5')

# loading encoder an scaler and encoder

with open('Geo_Encoder_Obj.pkl','rb') as file:
    geo_encoder = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder = pickle.load(file)

with open('Scaler_Obj.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
## converting the input data to data frame

input_df = pd.DataFrame([input_data])
```

```python
print(type(geo_encoder))
```

```python
# geo_encoded_data = geo_encoder.transform(input_data[['Geography']]).toarray()
geo_encoded = geo_encoder.transform(input_df[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded,columns=geo_encoder.get_feature_names_out())
```

```python
geo_encoded_df
```

```python
## Now concatinating geo_encoded_df
input_df = pd.concat([input_df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
input_df['Gender'] = label_encoder.transform(input_df['Gender'])
```

```python
input_df
```

```python
scaled_df = scaler.transform(input_df)
scaled_df
```

```python
prediction = model.predict(scaled_df)
```

```python
prediction_output= prediction[0][0]
print(prediction_output)
```

```python
if prediction_output>0.5:
    print('The Customer is likely to churn')
else:
    print('The Cutomer is not likely to churn')
```
